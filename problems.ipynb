## Theory / Background 

**SHA-256** is a **cryptographic hash function** from the **Secure Hash Standard**. Its job is to take any input and turn it into a **fixed 256-bit hash value**. This value is **one-way**, meaning you can't reverse it to get the original message. It's also **deterministic**, so the same input always gives the same output, and it's designed to **avoid collisions**, where two different inputs produce the same hash.

**SHA-256** was developed by the **National Institute of Standards and Technology (NIST)** and released in **2001** as part of the **Secure Hash Standard (FIPS PUB 180-2)**. It was introduced to replace older hash functions like **SHA-1**(which is now deprecated) and to provide a stronger, modern hashing method for security, digital signatures, and data integrity.

The algorithm works by splitting the message into **512-bit blocks** and running them through a series of bitwise operations like ``XOR``, ``AND``, **rotations**, and **shifts**. Functions such as ``Ch``, ``Maj``, and the ``Sigma`` functions help mix the bits so that even a 1-bit change in the input causes a completely different output. This is called the **avalanche effect**.

**SHA-256** is widely used for **checking data integrity**, **digital signatures**, and **general security tasks**.

## Functions Without NumPy (With Masking)

You can write all of the following **Secure Hash Standard helper functions** without numpy by relying entirely on **Python’s** built-in integers and masking them to **32 bits**. **Python** integers automaticallu grow beyond **32 bits**, so we just need to force them back into **32-bit** range using ``& 0xFFFFFFFF``. 

*Here is a NumPy free version of the Parity function as an example:*

````python
def parity(x,y,z):
    """
    Computes the parity function (Bitwise X0R) without the use of NumPy. 
    
    This was done to experiment with implementing Secure Hash Standard helper functions without relying on NumPy.
    Using masking to ensure 32-bit integer results.
    """
    
    return (x ^ y ^ z) & 0xFFFFFFFF
````

### Why This Works

- ``x ^ y ^ z`` > XORs the three values *(same behaviour as NumPy)*

- ``& 0xFFFFFFFF`` > masks the result so it behaves like an unsigned **32-bit integer.** Masking works because applying ``& 0xFFFFFFFF`` forces the result to fit into **exactly 32 bits**. The hexadecimal value ``0xFFFFFFFF`` is a number whose **binary form is 32 ones**, so when you perform a **bitwise AND** with it, all **bits outside the lower 32 are discarded**. **Python integers** normally expand beyond **32 bits** and don’t overflow like fixed-size machine integers, so this mask artificially enforces the same behaviour as a real **32-bit unsigned integer**. Any bitwise operation—such as ``x ^ y ^ z`` can then be “trimmed” to match **SHA-style 32-bit** logic simply by applying this mask.

It is important to note that this masking approach can be done for any of the helper function in the Secure Hash Standard to keep python integers within 32-bits. I just chose the Parity function as an example.

## Parity Function Implementation

- - - 

````python
def parity(x,y,z):
    """
    Computes the parity (bitwise XOR) of three 32-bit unsigned integers.
    
    Used as a logical function in SHA-1 algorithm. The parity is 1 if an 
    odd number of the three bits are 1, and 0 if an even number are 1.

    Parameters
    ----------
    x : int or np.uint32
        First integer input.
    y : int or np.uint32
        Second integer input.
    z : int or np.uint32
        Third integer input.

    Returns
    -------
    np.uint32
        The bitwise XOR of x, y, and z as a 32-bit unsigned integer.

    Examples
    --------
    >>> Parity(0b1010, 0b1100, 0b0011)
    np.uint32(5)
    """
    x = np.uint32(x)
    y = np.uint32(y)
    z = np.uint32(z)

    return np.uint32(x ^ y ^ z)
````

## Choose Function Implementation

---

````python
def choose(x,y,z):
    """
    Chooses bits from y or z based on the bits of x.
    
    If a bit in x is 1, the corresponding bit from y is chosen.
    If a bit in x is 0, the corresponding bit from z is chosen.
    This is a core logical function used in SHA-256 compression.

    Parameters
    ----------
    x : int or np.uint32
        Control input that determines which bits to choose.
    y : int or np.uint32
        First source input (chosen when x bit is 1).
    z : int or np.uint32
        Second source input (chosen when x bit is 0).

    Returns
    -------
    np.uint32
        Result of the expression (x & y) ^ (~x & z) as a 32-bit unsigned integer.

    Examples
    --------
    >>> Choose(0b1111, 0b1010, 0b0101)
    np.uint32(10)  # Binary: 0b1010 (all bits from y since x is all 1s)
    
    >>> Choose(0b0000, 0b1010, 0b0101)
    np.uint32(5)  # Binary: 0b0101 (all bits from z since x is all 0s)
    """
    x = np.uint32(x)
    y = np.uint32(y)
    z = np.uint32(z)

    return np.uint32((x & y) ^ ((~x) & z))
````

## Majority Function Documentation

- - -

````python
def majority(x,y,z):
    """
    Compute the bitwise majority function used in SHA-256.

    The majority function returns, for each bit position, the value that
    appears in at least two of the three 32-bit inputs. All inputs are
    converted to unsigned 32-bit integers to ensure correct wrapping and
    bitwise behaviour as defined in the Secure Hash Standard (FIPS 180-4).

    Parameters
    ----------
    x : int or np.uint32
        First input word.
    y : int or np.uint32
        Second input word.
    z : int or np.uint32
        Third input word.

    Returns
    -------
    np.uint32
        A 32-bit word where each bit is the majority bit of (x, y, z).

    Notes
    -----
    The function follows the SHA-256 definition:
    Maj(x, y, z) = (x & y) ^ (x & z) ^ (y & z)

    NumPy is used to enforce unsigned 32-bit arithmetic.

    Examples
    --------
    >>> Majority(0b1010, 0b1100, 0b0110)
    np.uint32(0b1010)
    """
    x = np.uint32(x)
    y = np.uint32(y)
    z = np.uint32(z)

    return np.uint32((x & y) ^ (x & z) ^ (y & z))
````

### Sigma0 Function Implementation

---

````python
def Sigma0(x):
    """
    Compute the SHA-256 Big Sigma 0 function.

    This function rotates the 32-bit input word three times (by 2, 13, and 22
    bits) and XORs the results together. The input is always converted to an
    unsigned 32-bit integer so the rotations wrap around correctly, following
    the SHA-256 rules in FIPS 180-4.

    Parameters
    ----------
    x : int or np.uint32
        The input 32-bit word.

    Returns
    -------
    np.uint32
        The result of ROTR^2(x) XOR ROTR^13(x) XOR ROTR^22(x).

    Notes
    -----
    Big Sigma 0 is one of the main mixing functions in SHA-256. NumPy is used
    to keep all operations within 32 bits.

    Examples
    --------
    >>> Sigma0(0b1010)
    np.uint32(…)
    """
    # Convert input to 32-bit unsigned integer
    x = np.uint32(x)
    
    # Rotate right by 2 bits: shift right 2, then move left bits wrap to left
    rotr2 = (x >> 2) | (x << (32 - 2))
    
    # Rotate right by 13 bits
    rotr13 = (x >> 13) | (x << (32 - 13))
    
    # Rotate right by 22 bits
    rotr22 = (x >> 22) | (x << (32 - 22))
    
    # XOR all three rotations together and ensure result is 32-bit
    return np.uint32(rotr2 ^ rotr13 ^ rotr22)
````

## Sigma1 Function Documentation

---

````python
def Sigma1(x):
    """
    Compute the SHA-256 Big Sigma 1 function.

    This function rotates the 32-bit input word three times (by 6, 11, and 25
    bits) and XORs the results together. The input is converted to an unsigned
    32-bit integer so the rotations wrap around correctly, following the rules
    of the SHA-256 standard.

    Parameters
    ----------
    x : int or np.uint32
        The input 32-bit word.

    Returns
    -------
    np.uint32
        The result of ROTR^6(x) XOR ROTR^11(x) XOR ROTR^25(x).

    Notes
    -----
    Big Sigma 1 is one of the main mixing functions in SHA-256. NumPy is used
    to keep the operations within 32 bits.

    Examples
    --------
    >>> Sigma1(0b1010)
    np.uint32(…)
    """
    # Convert input to 32-bit unsigned integer
    x = np.uint32(x)

    # Rotate right by 6 bits
    rotr6 = (x >> 6) | (x << (32 - 6))
    
    # Rotate right by 11 bits
    rotr11 = (x >> 11) | (x << (32 - 11))
    
    # Rotate right by 25 bits
    rotr25 = (x >> 25) | (x << (32 - 25))
    
    # XOR all three rotations together and ensure result is 32-bit
    return np.uint32(rotr6 ^ rotr11 ^ rotr25)
````

## sigma0 Function Implementation

---

````python
def sigma0(x):
    """
    Compute the SHA-256 small sigma 0 function.

    This function applies two right-rotations (by 7 and 18 bits) and one
    right-shift (by 3 bits) to the 32-bit input word, then XORs the results
    together. The input is converted to an unsigned 32-bit integer so all
    operations follow the 32-bit behaviour required by the SHA-256 standard.

    Parameters
    ----------
    x : int or np.uint32
        The input 32-bit word.

    Returns
    -------
    np.uint32
        The result of ROTR^7(x) XOR ROTR^18(x) XOR SHR^3(x).

    Notes
    -----
    Small sigma 0 is used in the message schedule part of SHA-256.
    NumPy ensures all shifts and rotations stay within 32 bits.

    Examples
    --------
    >>> sigma0(0b1010)
    np.uint32(…)
    """
    # Convert input to 32-bit unsigned integer
    x = np.uint32(x)

    # Rotate right by 7 bits
    rotr7 = (x >> 7) | (x << (32 - 7))
    
    # Rotate right by 18 bits
    rotr18 = (x >> 18) | (x << (32 - 18))
    
    # Right shift by 3 bits (no wraparound for shift, just lose the bits)
    shr3 = x >> 3
    
    # XOR all operations together and ensure result is 32-bit
    return np.uint32(rotr7 ^ rotr18 ^ shr3)
````

## sigma1 Function Implementation

---

````python
def sigma1(x):
    """
    Compute the SHA-256 small sigma 1 function.

    This function applies two right-rotations (by 17 and 19 bits) and one
    right-shift (by 10 bits) to the 32-bit input word, then XORs the results
    together. The input is converted to an unsigned 32-bit integer so the
    operations follow the correct 32-bit behaviour required by SHA-256.

    Parameters
    ----------
    x : int or np.uint32
        The input 32-bit word.

    Returns
    -------
    np.uint32
        The result of ROTR^17(x) XOR ROTR^19(x) XOR SHR^10(x).

    Notes
    -----
    Small sigma 1 is used in the message schedule step of SHA-256.
    NumPy ensures that all shifts and rotations stay within 32 bits.

    Examples
    --------
    >>> sigma1(0b1010)
    np.uint32(…)
    """
    # Convert input to 32-bit unsigned integer
    x = np.uint32(x)
    
    # Rotate right by 17 bits
    rotr17 = (x >> 17) | (x << (32 - 17))
    
    # Rotate right by 19 bits
    rotr19 = (x >> 19) | (x << (32 - 19))
    
    # Right shift by 10 bits (no wraparound)
    shr10 = x >> 10
    
    # XOR all operations together and ensure result is 32-bit
    return np.uint32(rotr17 ^ rotr19 ^ shr10)
````

# Primes Function Implementation

---

````python
def primes(n):
    """
    Generate the first n prime numbers.

    This function creates a list containing the first n prime numbers.
    It starts checking from 2 (the first prime) and tests each number
    for divisibility up to its square root using simple division.
    If a number has no divisors other than 1 and itself, it is added to
    the list of primes. The process continues until n prime numbers are found.

    Parameters
    ----------
    n : int
        The number of prime numbers to generate.

    Returns
    -------
    list of int
        A list containing the first n prime numbers.

    Notes
    -----
    - Uses a basic trial division method for simplicity.
    - NumPy is used only for the square root calculation (`np.sqrt`).
    - This method is efficient enough for small values such as n = 64,
      which is required for the SHA-256 constants.

    Examples
    --------
    >>> primes(5)
    [2, 3, 5, 7, 11]
    >>> primes(10)[-1]
    29
    """
    # Start with an empty list to store primes as we find them
    primes_list = []
    # Start checking from 2 (the first prime number)
    num = 2  

    # Keep going until we have found n prime numbers
    while len(primes_list) < n:
        # Check if num is divisible by any number from 2 to sqrt(num)
        # We only need to check up to sqrt because if num has a factor larger
        # than sqrt(num), it must also have a factor smaller than sqrt(num)
        for i in range(2, int(np.sqrt(num)) + 1):
            if num % i == 0:  # If divisible, num is not prime
                break
        else:
            # The else clause of a for loop runs only if the loop completes
            # without breaking, meaning num has no divisors - it's prime!
            primes_list.append(num)
        
        # Move to the next number to check
        num += 1

    return primes_list


# Step 1: Generate the first 64 primes (required for SHA-256 K constants)
prime_numbers = primes(64)

# Step 2: Calculate the cube root of each prime using NumPy
cube_roots = np.cbrt(prime_numbers)

# Step 3: Extract only the fractional part (everything after the decimal point)
# For example: if cube_root = 1.259921, fractional_part = 0.259921
fractional_parts = cube_roots - np.floor(cube_roots)

# Step 4: Convert the fractional part to a 32-bit integer
# Multiply by 2^32 to shift the fractional bits into the integer range
constants = np.uint32(fractional_parts * (2**32))

# Step 5: Display results in hexadecimal format for comparison with SHA-256 standard
for i, value in enumerate(constants):
    print(f"{i+1:02d}. Prime: {prime_numbers[i]:3d}  = Constant: 0x{value:08X}")
````

## My Understanding

In this problem, I need to make a **generator function** called ``block_parse(msg)`` that splits a message into **512-bit blocks and adds the correct SHA-256 padding** at the end. A **generator function** in **Python** is a special type of function that uses the ``yield`` keyword instead of ``return``. This means it **doesn’t give all results at once** — it **produces one value at a time**, which is useful here because the message is processed in separate blocks.

According to sections **5.1.1** and **5.2.1** of the **Secure Hash Standard (FIPS 180-4)**, a **message in SHA-256 must be padded before hashing**. **Padding means extra bits are added to make the message length fit exactly into 512-bit blocks**. The rule says you always **add one 1 bit after the message**, followed by **enough 0s to make the total length 64 bits short of a multiple of 512**. Then, you add the original message length (in bits) as a **64-bit value at the very end**.

So, this function will take a message, pad it correctly, and then **yield each 512-bit chunk one at a time**. This process prepares the data for the main hashing steps in **SHA-256**. It’s a good way to learn how the algorithm structures its input before the actual hash calculations begin.

## Theory / Background

In **SHA-256**, the message can’t be hashed directly in one go. Instead, it is **split into fixed-size 512-bit blocks**. Real messages almost never fit perfectly into these block sizes, so we need padding. Padding is a way of adding extra bits in a standard, predictable way so that:

- Every possible original message has **one unique padded form**.

- The **final padded message length is a multiple of 512 bits**.

*For SHA-256, the padding rule is always the same:*

1. Start with the original message (in bits).

2. Append a single **1** bit.

3. **Append 0 bits until the length is 64 bits short of a multiple of 512**.

4. Finally, **append the original message length (in bits) as a 64-bit number**.

This idea comes from older hash function designs (like **MD4/MD5** and **SHA-1**) and was standardised in the **Secure Hash Standard (FIPS 180) by NIST**, based on designs originally created by the **NSA**. The goal is to make the padding rule public, simple, and safe, so there is no hidden trick in how messages are prepared.

On the **Python** side, a **generator function** is a function that can produce a sequence of values one at a time, instead of building everything in memory at once. Generators were introduced into **Python** by **Guido van Rossum** and others via **PEP 255**, first appearing in **Python 2.2**. They use the **yield** keyword to *“pause”* and *“resume”* execution.

In this problem, using a generator fits nicely: ``block_parse(msg)`` can yield one padded **512-bit block at a time**, which matches how **SHA-256** itself processes data internally—block by block—in a streaming fashion, rather than needing the whole padded message as a single big object.

## Approach

For this problem, my goal was to write a **generator** that prepares a message exactly the way **SHA-256** expects before hashing. I broke the task into simple steps that follow the padding rules from the Secure Hash Standard.

First, I calculated the original length of the message in bits, because **SHA-256** must store that at the very end. Then I added the required ``1`` bit, which is represented as the **byte 0x80**. After that, I **added zero bytes until the message length was exactly 64 bits away from a multiple of 512 bits**, which is what the standard requires.

Once that condition was met, I added the **64-bit length field (big-endian)**, which completes the padding. At that point, the padded message is **guaranteed to be a clean multiple of 512 bits**, so I simply sliced it into **64-byte blocks** and used ``yield`` to produce one block at a time. Using ``yield`` keeps the function efficient and makes it easier to process long messages without storing every block at once.

Overall, I followed the **SHA-256** rules step by step and wrote the generator in the simplest way possible, focusing on clarity and correctness rather than complexity.

# Padding Function Implementation

- - -

````python
def block_parse(msg):
    """
    Split a message into 512-bit blocks with correct SHA-256 padding.

    This function takes a bytes message and pads it according to the
    SHA-256 rules. It adds the required '1' bit, then enough zeros, and
    finally the original message length in 64 bits. After padding, it
    produces each 512-bit (64-byte) block one at a time using `yield`,
    which makes the function a generator.

    Parameters
    ----------
    msg : bytes
        The original message to be padded and split into blocks.

    Returns
    -------
    bytes
        Each call to `yield` returns the next 64-byte block of the padded
        message.

    Notes
    -----
    - Follows the padding rules from the Secure Hash Standard (FIPS 180-4).
    - The message length is stored as a 64-bit big-endian value.
    - Padding ensures the final message size is a multiple of 512 bits.

    Examples
    --------
    >>> list(block_parse(b"abc"))
    [b'...64 bytes...', ...]
    """
    
    # Store the original message length in bits (multiply by 8 since len() gives bytes)
    original_len_bits = len(msg) * 8

    # Step 1: Append the '1' bit (represented as 0x80 = 10000000 in binary)
    padded = msg + b"\x80"

    # Step 2: Append zeros until we're 64 bits short of a full 512-bit block
    # 512 bits = 64 bytes, so 448 bits = 56 bytes (need room for 8-byte length)
    while (len(padded) * 8) % 512 != 448:
        padded += b"\x00"

    # Step 3: Append the original message length as a 64-bit big-endian integer
    # This goes at the end of the last block
    padded += original_len_bits.to_bytes(8, "big")

    # Step 4: Yield each 512-bit (64-byte) block to the caller
    # Using a generator saves memory compared to creating the whole list at once
    for i in range(0, len(padded), 64):
        yield padded[i:i+64]
````

## Theory / Background

Hashing has existed in computing for decades and is used to turn any input into a **fixed-size output**. The general idea is that you take some data, break it into pieces, and then **update a running value each time you process a new piece**. This running value is often called the **current state**. When you combine this state with the next chunk of data, you get a new state, and this repeats until the whole message is processed.

The concept of updating a state step-by-step came from early work in computer science on message processing and error-checking, and later became a core idea in many hashing systems. The important part is that each block of the message affects the next state, so the final output depends on every block before it.

In this problem, the function you write will follow this same pattern: take the current value, mix it with the next block, and output an updated value. For this problem, my approach was to follow the structure that **Section 6.2.2** describes, but to write it in a way that is as clear and simple as possible. The goal was to *take the current hash values and mix them with the next message block*, step by step, until a new hash state is produced.

# Hash Function Implementation

````python

# SHA-256 Round Constants (K values)
# These are the first 32 bits of the fractional parts of the cube roots of the first 64 primes
K = np.array([
    0x428a2f98, 0x71374491, 0xb5c0fbcf, 0xe9b5dba5, 0x3956c25b, 0x59f111f1,
    0x923f82a4, 0xab1c5ed5, 0xd807aa98, 0x12835b01, 0x243185be, 0x550c7dc3,
    0x72be5d74, 0x80deb1fe, 0x9bdc06a7, 0xc19bf174, 0xe49b69c1, 0xefbe4786,
    0x0fc19dc6, 0x240ca1cc, 0x2de92c6f, 0x4a7484aa, 0x5cb0a9dc, 0x76f988da,
    0x983e5152, 0xa831c66d, 0xb00327c8, 0xbf597fc7, 0xc6e00bf3, 0xd5a79147,
    0x06ca6351, 0x14292967, 0x27b70a85, 0x2e1b2138, 0x4d2c6dfc, 0x53380d13,
    0x650a7354, 0x766a0abb, 0x81c2c92e, 0x92722c85, 0xa2bfe8a1, 0xa81a664b,
    0xc24b8b70, 0xc76c51a3, 0xd192e819, 0xd6990624, 0xf40e3585, 0x106aa070,
    0x19a4c116, 0x1e376c08, 0x2748774c, 0x34b0bcb5, 0x391c0cb3, 0x4ed8aa4a,
    0x5b9cca4f, 0x682e6ff3, 0x748f82ee, 0x78a5636f, 0x84c87814, 0x8cc70208,
    0x90befffa, 0xa4506ceb, 0xbef9a3f7, 0xc67178f2
], dtype=np.uint32)


def hash(current, block):
    """
    Compute one SHA-256 compression function iteration.
    
    Takes the current hash state and a 512-bit (64-byte) message block,
    and returns the updated hash state after processing through 64 rounds.
    """

    # Step 1: Expand the 16 message words into 64 working variables (W array)
    W = np.zeros(64, dtype=np.uint32)

    # Copy the first 16 words directly from the message block (each word is 4 bytes)
    for i in range(16):
        W[i] = np.uint32(int.from_bytes(block[i*4:(i+1)*4], 'big'))

    # Generate words 16-63 using the sigma functions
    # Each new word depends on previous words mixed with sigma0 and sigma1
    for i in range(16, 64):
        s1 = sigma1(W[i-2])          # Apply sigma1 to word 2 positions back
        s0 = sigma0(W[i-15])         # Apply sigma0 to word 15 positions back
        # Add them all together with bitwise AND to keep result in 32-bit range
        temp = (int(s1) + int(W[i-7]) + int(s0) + int(W[i-16])) & 0xffffffff
        W[i] = np.uint32(temp)

    
    # Step 2: Initialize working variables with current hash state
    # These are the 8 32-bit values that represent the current hash
    a, b, c, d, e, f, g, h = [int(x) for x in current]

    
    # Step 3: Main SHA-256 compression loop (64 rounds)
    for i in range(64):
        # Calculate T1: sum of h + Sigma1(e) + Ch(e,f,g) + K[i] + W[i]
        # This combines the current hash value with new mixing
        T1 = (
            h +
            int(Sigma1(e)) +
            int(choose(e, f, g)) +
            int(K[i]) +
            int(W[i])
        ) & 0xffffffff  # Keep result in 32-bit range

        # Calculate T2: sum of Sigma0(a) + Maj(a,b,c)
        # This provides additional mixing based on the top hash values
        T2 = (int(Sigma0(a)) + int(majority(a, b, c))) & 0xffffffff

        
        # Rotate the 8 working variables
        # Move each variable one position, calculate new values from T1 and T2
        h = g           # h gets previous g
        g = f           # g gets previous f
        f = e           # f gets previous e
        e = (d + T1) & 0xffffffff  # e gets previous d plus T1
        d = c           # d gets previous c
        c = b           # c gets previous b
        b = a           # b gets previous a
        a = (T1 + T2) & 0xffffffff  # a gets sum of both T values

    
    # Step 4: Add the working variables back to the original hash state
    # This completes one compression function round
    return [
        np.uint32((int(current[0]) + a) & 0xffffffff),
        np.uint32((int(current[1]) + b) & 0xffffffff),
        np.uint32((int(current[2]) + c) & 0xffffffff),
        np.uint32((int(current[3]) + d) & 0xffffffff),
        np.uint32((int(current[4]) + e) & 0xffffffff),
        np.uint32((int(current[5]) + f) & 0xffffffff),
        np.uint32((int(current[6]) + g) & 0xffffffff),
        np.uint32((int(current[7]) + h) & 0xffffffff)
    ]
````

## Theory & Background

A **hash function** is a process that takes an input—such as a password—and converts it into a fixed, scrambled output known as a **hash**. The key idea is that the hash does not reveal the original password directly. Hashing is commonly used when storing passwords so that even if someone gains access to the database, they cannot immediately see the real passwords.

However, hashing becomes weak when the input passwords themselves are simple or very common. If a user picks an easy password like *“password”* or *“123456”*, attackers can easily guess or look them up, because these same inputs always produce the same hash. Large online collections of known password hashes exist, so anyone can copy a hash into a lookup tool and instantly find the original password if it is common enough. This is not breaking the hash function—it is exploiting predictable user behaviour.

Before hashing, passwords are treated as text, and this requires a **character encoding**. The question mentions **UTF-8**, which is the most widely used **encoding** system today. **UTF-8** was designed to represent every character in the **Unicode standard** while staying compatible with older **ASCII systems**. It uses **1 byte** for simple characters (like typical English letters) and more bytes for symbols or characters from other languages. This matters because the exact bytes produced by **UTF-8** are what get passed into the hashing function; different encodings would produce different hashes even for the same visible text. **UTF-8** ensures a consistent and universal way to turn text into bytes before hashing.

Overall, the background to this question is that **hashing** is only as strong as the passwords being hashed. If passwords are short, predictable, or extremely common, **hashes** offer very little protection, because attackers can match the hash to known values. Understanding hashing, how passwords are encoded with *UTF-8*, and why simple passwords are vulnerable is essential to explaining how these given hashes can be reversed.

## Passwords Function Implementation

- - -


````python
def sha256_utf8(text):
    """
    Compute the SHA-256 hash of a given string encoded in UTF-8.

    This function takes an input string, encodes it using UTF-8,
    and computes its SHA-256 digest using Python's hashlib library.
    The output is returned as a lowercase hexadecimal string, matching
    the standard representation used in password hashing and security
    applications.

    Parameters
    ----------
    text : str
        The input string to be hashed. It will be encoded to UTF-8
        before the hashing process.

    Returns
    -------
    str
        A 64-character hexadecimal string representing the SHA-256
        digest of the UTF-8 encoded input.

    Notes
    -----
    - Follows the SHA-256 hashing procedure defined in FIPS 180-4.
    - UTF-8 encoding ensures consistent byte representation of the text.
    - hashlib handles internal bitwise operations and 32-bit word behaviour.

    Examples
    --------
    >>> sha256_utf8("password")
    '5e884898da28047151d0e56f8dc6292773603d0d6aabbdd62a11ef721d1542d8'
    >>> sha256_utf8("hello")
    '2cf24dba5fb0a30e26e83b2ac5b9e29e1b161e5c1fa7425e73043362938b9824'
    """
    # Step 1: Encode the text string to UTF-8 bytes
    # UTF-8 is the standard encoding that converts text into raw bytes
    utf8_bytes = text.encode("utf-8")
    
    # Step 2: Hash the bytes using SHA-256 algorithm
    # hashlib.sha256() creates a hash object from the input bytes
    hash_object = hashlib.sha256(utf8_bytes)
    
    # Step 3: Convert the hash to a hexadecimal string representation
    # hexdigest() gives us the familiar 64-character hex string (256 bits / 4 bits per hex char)
    hex_string = hash_object.hexdigest()
    
    return hex_string

# Step 1: Create a list of the most common passwords found in real data breaches
# These are based on statistics from password leak analysis and are extremely common
common_passwords = np.array([
    "123456",
    "admin",
    "12345678",
    "123456789",
    "1234",
    "12345",
    "password",
    "123",
    "Aa123456",
    "1234567890",
    "UNKNOWN",
    "Password",
    "123321",
    "1234567",
    "welcome",
    "password1",
    "P@ssw0rd",
    "qwerty",
    "iloveyou",
    "123123"
])

# Step 2: Define the target hashes we want to crack
# Each hash maps to None initially (we'll fill in the password when we find it)
target_hashes = {
    "5e884898da28047151d0e56f8dc6292773603d0d6aabbdd62a11ef721d1542d8": None,
    "873ac9ffea4dd04fa719e8920cd6938f0c23cd678af330939cff53c3d2855f34": None,
    "b03ddf3ca2e714a6548e7495e2a03f5e824eaac9837cd7f159c67b90fb4b7342": None
}

# Step 3: Perform the dictionary attack - try each common password
# This is the core of the attack: hash each password and see if it matches any target
for pwd in common_passwords:
    # Hash the current password using UTF-8 encoding
    hashed = sha256_utf8(pwd)
    
    # Check if this hash matches any of our target hashes
    # If it does, we found the password!
    if hashed in target_hashes:
        target_hashes[hashed] = pwd

# Step 4: Print the results - show which hashes we cracked and their passwords
# None values mean we didn't find that password in our dictionary
for h, pwd in target_hashes.items():
    print(f"{h}  -->  {pwd}")
````
